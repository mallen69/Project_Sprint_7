```r
hart_tb = hart_tb %>% select(ROUTE, Direction_Code, ETHNICITY_Code, ETHNICITY_Text, GENDER_Text, TotalTransfers, `PURPOSE Text`)
```

```r
korean_riders = hart_tb %>% filter(`ETHNICITY_Code` == 4)
korean_riders %>% count
korean_riders %>% head
```

```r
rider_slice = hart_tb %>% slice(100:150)
rider_slice %>% count
rider_slice %>% head
```

```r
rider_sample_1_percent = hart_tb %>% sample_frac(.01)
rider_sample_1_percent %>% count
rider_sample_1_percent %>% head
```

```r
rider_sample_100 = hart_tb %>% sample_n(100)
rider_sample_100 %>% count
rider_sample_100 %>% head
```

```r
routes = hart_tb %>% group_by(ROUTE) %>% count
routes
```

```r
routes = hart_tb %>% group_by(ROUTE) %>% summarise(count = n(), n_distinct(ETHNICITY_Code),max(ETHNICITY_Code), sum(TotalTransfers))
routes
```
